# Pre Processing Data

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels =[]
train_samples = []
```

# Example Data:
An experimental drug was tested on indiviuals from age 13 to 100
The trial had 2100 participants.Half were under 65 year old.Half were above 65 year old.
95% of patients 65 or older experience side effects.
95% of patients 65 or older experience no side effects.

```python
for i in range (1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range (50):
    
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)

    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
#print raw data
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array (train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples= scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print (i)
```

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(16, input_shape=(1,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics =['accuracy'])
```

```python
model.fit(scaled_train_samples, train_labels, batch_size=10, epochs=20, shuffle=True, verbose=2)
```
